```python
import cv2
import matplotlib.pyplot as plt

#ie aktar
einstein=cv2.imread("eis.jpeg",0)
plt.figure(),plt.imshow(einstein,cmap="gray"),plt.axis("off")

good=cv2.imread("good.jpeg",0)
plt.figure(),plt.imshow(good,cmap="gray"),plt.axis("off")


#snflandrc
#cascade daha nce eitilmi snflandrcalardr
face_cascade=cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
face_rect=face_cascade.detectMultiScale(einstein)

for(x,y,w,h) in face_rect:
  cv2.rectangle(einstein,(x,y),(x+w,y+h),(255,255,255),10)
plt.figure(),plt.imshow(einstein,cmap="gray"),plt.axis("off")

face_rect=face_cascade.detectMultiScale(good,minNeighbors=5)
for(x,y,w,h) in face_rect:
  cv2.rectangle(good,(x,y),(x+w,y+h),(255,255,255),5)
plt.figure(),plt.imshow(good,cmap="gray"),plt.axis("off")


#video kameradan yz alglama
#cap=cv2.VideoCapture(0)

#while True:
 # ret, frame=cap.read()
  #if ret:
   # face_rect=face_cascade.detectMultiScale(frame,minNeighbors=5)
    #for(x,y,w,h) in face_rect:
     #  cv2.rectangle(frame,(x,y),(x+w,y+h),(255,255,255),5)
      # cv2.imshow("face decect",frame)
  #if cv2.waitKey(1)& 0xFF==ord("q"): break

#cap.release()
#cv2.destroyAllWindows()














```
